```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

output_dir = "../output/"
```

```python
result = pd.read_csv(os.path.join(output_dir, "TrainingResult.csv"))
```

```python
result
```

```python
names = ["ConvLSTM","GCNLSTM","T-GCNLSTM","STGCN"]
labels = ["CNN-LSTM","GCN-LSTM","T-GCN-LSTM","ST-GCN"]
```

```python
loss_colnames = []
time_colnames = []

for name in names:
    loss_colnames.append(f"{name}_loss")
    time_colnames.append(f"{name}_time")
```

```python
fig, ax = plt.subplots(figsize = (7,5))

boxprops = dict(linewidth=1.7, color='black')
flierprops = dict(marker='o', markerfacecolor='green', markersize=12,
                  markeredgecolor='none')
medianprops = dict(linewidth=1.7, color='blue')
meanlineprops = dict(linestyle='--', linewidth=1.5, color='purple')

#ax.boxplot(result[loss_colnames], widths = 0.4, boxprops = boxprops, flierprops = flierprops, medianprops = medianprops,
#          meanprops = meanlineprops)
ax.boxplot(result[loss_colnames], widths = 0.4, meanline=True, boxprops = boxprops, medianprops = medianprops)

ax.set_ylabel("BCE Loss", fontsize = 12)
ax.set_xlabel("Methods", fontsize = 12, labelpad=10)
ax.set_xticks([1,2,3,4], labels)
ax.set_ylim(0,43)
ax.set_title("Testing Loss of Wildfire Prediction through Different Methods", y = 1.04, fontsize = 14)

#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)

fig.savefig("../fig/TestLoss.png", dpi = 400)
```

```python
fig, ax = plt.subplots(figsize = (7,5))

boxprops = dict(linewidth=1.7, color='black')
flierprops = dict(marker='o', markerfacecolor='green', markersize=12,
                  markeredgecolor='none')
medianprops = dict(linewidth=1.7, color='red')
meanlineprops = dict(linestyle='--', linewidth=1.5, color='purple')

#ax.boxplot(result[loss_colnames], widths = 0.4, boxprops = boxprops, flierprops = flierprops, medianprops = medianprops,
#          meanprops = meanlineprops)
ax.boxplot(result[time_colnames], widths = 0.4, meanline=True, boxprops = boxprops, medianprops = medianprops)

ax.set_ylabel("Time Elapse (s)", fontsize = 12)
ax.set_xlabel("Methods", fontsize = 12, labelpad=10)
ax.set_xticks([1,2,3,4], labels)
ax.set_ylim(0,2100)
ax.set_title("Training Time of Different Methods", y = 1.04, fontsize = 14)

#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)

fig.savefig("../fig/TimeElapse.png", dpi = 400)
```
